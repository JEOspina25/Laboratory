# Reflectancia Al

Se toman los datos de reflectancia del aluminio y se ajusta una función de la forma:

$$ R_{\lambda} = \frac{S_{\lambda}- D_{\lambda}}{R_{\lambda}- D_{\lambda}} $$

Donde $R$ es la reflectancia y $\lambda$ es la longitud de onda en nm.

```python
df_AlD = pd.read_csv('clean_data/AL.Master.dark.csv')
df_AlR = pd.read_csv('clean_data/AL.Master.reference.csv')
df_AlS = pd.read_csv('clean_data/AL.Master.sample.csv')

dfnum = (df_AlS['Intensity'] - df_AlD['Intensity'])*100/(df_AlR['Intensity'] - df_AlD['Intensity'])
plt.title('Al')
plt.plot(df_AlS['Wavelength'],dfnum, label = 'Al')
plt.xlabel('Longitud de onda (nm)')
plt.ylabel('% de reflectancia')
plt.ylim(0,100)
```

# Indice de refracción y coeficiente de extinción

En las gráficas a continuación se presentan los resultados obtenidos asociados a $n$ y $k$, a partir de la medición de la reflectancia y con las relaciones de Kramers - Kronig.

```python
def plot(dark,reference,sample,material):

    """
    Esta función nos permite obtener la gráfica correspondiente a las curvas
    experimentales de n y k para cada material estudiado
    Nos basamos en el código provisto de las relaciones Kramers-Kronig

    Inputs:
    dark: señal oscura de la referencia
    reference: señal de la referencia
    sample: señal de la muestra
    material: material del cual son los datos
    """

    #Usamos pandas para cargar los espectros tomados con el Ocean Optics 2000
    #lim y lim2 nos permite acotar el rango del espectro a tomar. Las longuitudes de onda donde la señal de referencia es muy baja deben despreciarse, ya tendran un gran error en la reflectancia.
    lim=300
    lim2=200
    #Window size determina un tamaño de filtro para suavizar el espectro. Debe ser un valor par.
    Window_size=201
    #Se carga la señal oscura de la referencia
    f = pd.read_csv (dark, sep='\t',skiprows=14, skipfooter=1, engine='python')
    Ir = np.array(f)[lim:-lim2,1]
    wld = np.array(f)[lim:-lim2,0]
    #Se carga la señal de la referencia
    f = pd.read_csv (reference, sep='\t',skiprows=14, skipfooter=1, engine='python')
    Is = np.array(f)[lim:-lim2,1]
    #Se carga la señal oscura de la muestra
    f = pd.read_csv (dark, sep='\t',skiprows=14, skipfooter=1, engine='python')
    Ir1 = np.array(f)[lim:-lim2,1]
    # Se carga la señal de la muestra
    f = pd.read_csv (sample, sep='\t',skiprows=14, skipfooter=1, engine='python')
    Is1 = np.array(f)[lim:-lim2,1]
    Is1=Is1
    Is=Is
    # Aquí se filtra el espectro y la referencia
    Is1=medfilt(medfilt(Is1, kernel_size=Window_size))
    Is=medfilt(medfilt(Is, kernel_size=Window_size))
    # Se calcula la reflectancia y grafica
    R1=Is1/Is


    # Finalmente, calculamos la fase del coeficiente de reflexion a partir de la reflectancia
    wn2d = 1/wld[::-1]
    R2=R1
    lnr=np.log(R2)/2
    # Método McClaurin
    dn = np.zeros(len(wn2d))
    h = wn2d[1]-wn2d[0]
    for ii in range(len(wn2d)):
        if(ii % 2 == 0 ):
            f = 0
            for jj in range(1,len(wn2d),2):
                f = f + wn2d[ii]*lnr[jj]/(wn2d[jj]**2 - wn2d[ii]**2)
        else:
            f = 0
            for jj in range(0,len(wn2d),2):
                f = f + wn2d[ii]*lnr[jj]/(wn2d[jj]**2 - wn2d[ii]**2)
        dn[ii] = 2*2*h/np.pi*f
        phaser = -dn

    nn=(1-R2**2)/(1+R2**2-2*R2*np.cos(phaser))
    kk=2*R2*np.sin(phaser)/(1+R2**2-2*R2*np.cos(phaser))

    plt.figure(figsize=(16,6))
    plt.plot(wld,nn,'r-',label='n experimental')
    plt.plot(wld,kk,'g-',label='k experimental')
    plt.title("Valores de n y k en función de la longitud de onda para " + material)
    plt.xlabel(r'$\lambda$ (nm)')
    plt.legend()
    plt.grid()
    plt.show()
    return
```

```python
plot('Datos/AL.Master.dark','Datos/AL.Master.reference','Datos/AL.Master.sample','Al')
```
